````python

import qiskit
import qiskit_aer
````

````python
# can you create a quantum circuit?
from qiskit import QuantumCircuit
qc = QuantumCircuit(2)
qc.h(0) 
````

````python
# Create a 5-qubit quantum circuit
qc5 = QuantumCircuit(5)

# Add some quantum gates
qc5.h(0)  # Hadamard gate on qubit 0
qc5.cx(0, 1)  # CNOT gate
qc5.cx(1, 2)  # CNOT gate
qc5.h(3)  # Hadamard gate on qubit 3
qc5.cx(3, 4)  # CNOT gate

# Display the circuit
print(qc5)
qc5.draw('mpl')
````

# Documentation: 5-qubit circuit (qc5)

This cell creates a 5-qubit `QuantumCircuit` named `qc5` and applies the following gates:
- `qc5.h(0)`: Hadamard on qubit 0 (puts qubit 0 into superposition).
- `qc5.cx(0, 1)`: CNOT with control qubit 0 and target qubit 1 (entangles 0 and 1).
- `qc5.cx(1, 2)`: CNOT with control qubit 1 and target qubit 2 (extends entanglement).
- `qc5.h(3)`: Hadamard on qubit 3 (creates superposition on qubit 3).
- `qc5.cx(3, 4)`: CNOT with control qubit 3 and target qubit 4 (entangles 3 and 4).

Usage notes:
- `print(qc5)` shows a text summary of the circuit.
- `qc5.draw('mpl')` renders a matplotlib diagram (requires `matplotlib`).
- To simulate the circuit, use `qiskit_aer`'s `AerSimulator` and run with `execute` or assemble+run.

Example (simulate):
```python
from qiskit import transpile
from qiskit_aer import AerSimulator
sim = AerSimulator()
tqc = transpile(qc5, sim)
result = sim.run(tqc).result()
print(result.get_counts())
```
